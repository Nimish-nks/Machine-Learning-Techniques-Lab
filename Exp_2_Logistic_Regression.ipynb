#Experiment-2: Logistic Regression
To create and test Logistic regression model on 'customer churn' dataset. Your model should predict how many customers will stay in the company.

Dataset: ChurnData.csv
#Import necessary libraries

```python
import pandas as pd
```

#Get data

```python
df=pd.read_csv('/content/ChurnData.csv')
df.head()
```

#Explore Data

Let's see datatypes of features

```python
df.info()
```

Let's select specific features and create new dataframe 'churn_df'. Datatype of target variable 'churn' should be of integer type for logistic regression.


```python
churn_df = df[['tenure','age','address','income','ed','employ','equip','callcard','wireless','churn']]
churn_df['churn']=churn_df['churn'].astype('int')
churn_df.head()
```

# Train Test split

```python
from sklearn.model_selection import train_test_split
train_set, test_set = train_test_split(churn_df, test_size=0.2, random_state=42)
train_set.head()
```

```python
X_train = train_set[['tenure','age','address','income','ed','employ','equip','callcard','wireless']]
Y_train=train_set['churn']
Y_test = test_set['churn']
X_test = test_set[['tenure','age','address','income','ed','employ','equip','callcard','wireless']]
```

In logistic regression, one always has to normalize feature values

```python
for col in X_train.columns.to_list():
    X_train[col]=(X_train[col]-X_train[col].mean())/X_train[col].std()
for col in X_test.columns.to_list():
    X_test[col]=(X_test[col]-X_test[col].mean())/X_test[col].std()
```

#Training

This function  LogisticRegression and can use different numerical optimizers to find parameters, including â€˜newton-cgâ€™, â€˜lbfgsâ€™, â€˜liblinearâ€™, â€˜sagâ€™, â€˜sagaâ€™ solvers. You can find extensive information about the pros and cons of these optimizers if you search it in the internet.

The version of Logistic Regression in Scikit-learn, support regularization. Regularization is a technique used to solve the overfitting problem of machine learning models. C parameter indicates inverse of regularization strength which must be a positive float. Smaller values specify stronger regularization.


```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix
LR = LogisticRegression(C=0.01, solver='liblinear').fit(X_train,Y_train)
LR
```

# Predict target variable on Test Set

Let's predict the churn values for  our test set

```python
yhat = LR.predict(X_test)
yhat
```

Let's see predicted probabilities

```python
yhat_prob = LR.predict_proba(X_test)
yhat_prob
```

#Evaluation

Jaccard index:  If the entire set of predicted labels for a sample strictly matches with the true set of labels, then the Jaccard index is 1.0; otherwise it is 0.0.

```python
from sklearn.metrics import jaccard_score
jaccard_score(Y_test, yhat, pos_label=0)
```

#Confusion Matrix


```python
cnf_matrix = confusion_matrix(Y_test, yhat, labels=[1,0])
cnf_matrix

```
